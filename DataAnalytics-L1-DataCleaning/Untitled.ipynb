## Objective

The objective of this project is to clean a raw dataset by identifying and handling missing values, duplicate records, inconsistent formatting, incorrect data types, and outliers. The cleaned dataset will be analysis-ready and saved as a new CSV file.


```python
# Import required libraries

import pandas as pd
import numpy as np

# Display all columns when printing DataFrames
pd.set_option("display.max_columns", None)
```

```python
# Load dataset
df = pd.read_csv("dirty_dataset.csv")

# Display first 5 rows
df.head()
```

## Handling Missing Values

Missing values were identified in the dataset. Appropriate techniques such as median imputation, mode imputation, and column removal were used depending on the nature of each column.

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Fill missing Fare with median (if any)
if df["Fare"].isnull().sum() > 0:
    df["Fare"] = df["Fare"].fillna(df["Fare"].median())

# Drop Cabin column because it contains too many missing values
if "Cabin" in df.columns:
    df.drop(columns=["Cabin"], inplace=True)

print("Missing values handled successfully.")
```

## Removing Duplicate Records

Duplicate rows were checked and removed to ensure data integrity.

```python
duplicates_before = df.duplicated().sum()

df.drop_duplicates(inplace=True)

duplicates_after = df.duplicated().sum()

print("Duplicates Before:", duplicates_before)
print("Duplicates After:", duplicates_after)
```

## Standardization

Text values were standardized to maintain consistency across the dataset.

```python
# Standardize Sex column
df["Sex"] = df["Sex"].str.lower().str.strip()

df["Sex"] = df["Sex"].replace({
    "male": "Male",
    "female": "Female"
})

# Standardize Embarked column
df["Embarked"] = df["Embarked"].str.upper()

print(df["Sex"].unique())
print(df["Embarked"].unique())
```

## Outlier Detection

The IQR (Interquartile Range) method was used to detect outliers in numerical columns. Outliers were retained because they may represent genuine passenger information.

```python
numeric_columns = ["Age", "Fare"]

for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} Outliers: {len(outliers)}")
```

## Data Type Correction

The data types were verified and corrected where necessary to ensure consistency for further analysis.

```python
# Convert PassengerId to string
df["PassengerId"] = df["PassengerId"].astype(str)

# Verify data types
df.dtypes
```

## Before vs After Summary

```python
# Reload original dataset
original_df = pd.read_csv("dirty_dataset.csv")

summary = pd.DataFrame({
    "Metric": [
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Rows"
    ],
    "Before Cleaning": [
        original_df.shape[0],
        original_df.shape[1],
        original_df.isnull().sum().sum(),
        original_df.duplicated().sum()
    ],
    "After Cleaning": [
        df.shape[0],
        df.shape[1],
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ]
})

summary
```
